```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "subhalo_242805_z20.csv"
#Cols = ['id', 'x_ics', 'y_ics', 'z_ics', 'x_z0', 'y_z0', 'z_z0', 'potential_z0']
data = np.loadtxt(path, delimiter=',', skiprows=1)
x_ics = data[:, 1]
y_ics = data[:, 2]
z_ics = data[:, 3]
x_z0 = data[:, 4]
y_z0 = data[:, 5]
z_z0 = data[:, 6]
potential_z0 = data[:, 7]
```

```python
#graficar la distribucion del potencial de las particulas del subhalo
plt.figure(figsize=(8, 5))
plt.hist(potential_z0, bins=50, color='darkred')
plt.title('Distribucin del Potencial de las Partculas del Subhalo')
plt.xlabel('Potencial')
plt.ylabel('Nmero de Partculas')
plt.grid(alpha=0.5)
plt.show()
```

```python
#proyeccion 2D de las posiciones iniciales y finales de las particulas de un subhalo
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_ics, y_ics, s=1, c='b')
plt.title('Proyeccin 2D de Posiciones Iniciales')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 35000)
plt.ylim(0, 35000)
plt.grid()
plt.subplot(1, 2, 2)
plt.scatter(x_z0, y_z0, s=1, c='b')
plt.title('Proyeccin 2D de Posiciones Finales')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.xlim(0, 35000)
plt.ylim(0, 35000)
```

```python
#proyeccion 2D de las posiciones iniciales y finales de las particulas de un subhalo
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(x_ics, y_ics, s=1, c='b')
plt.title('Proyeccin 2D de Posiciones Iniciales')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.scatter(x_z0, y_z0, s=1, c='b')
plt.title('Proyeccin 2D de Posiciones Finales')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.axis('equal')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Posiciones iniciales
sc1 = ax1.scatter(x_ics, y_ics, s=1, c=np.abs(potential_z0), cmap='inferno')
ax1.set_title('Proyeccin 2D de Posiciones Iniciales')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid()
ax1.axis('equal')
plt.colorbar(sc1, ax=ax1, label='Potencial')

# Posiciones finales
sc2 = ax2.scatter(x_z0, y_z0, s=1, c=np.abs(potential_z0), cmap='inferno')
ax2.set_title('Proyeccin 2D de Posiciones Finales')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid()
ax2.axis('equal')
plt.colorbar(sc2, ax=ax2, label='Potencial')

plt.tight_layout()
plt.show()
```
