```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
halo3_d_z0 = pd.read_csv('../../results/Halos_50_3_z0.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_d_z0d1 = pd.read_csv('../../results/Halos_50_3_z0d1.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_d_z0d24 = pd.read_csv('../../results/Halos_50_3_z0d24.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_d_z0d5 = pd.read_csv('../../results/Halos_50_3_z0d5.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_d_z0d7 = pd.read_csv('../../results/Halos_50_3_z0d7.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_d_z1d25 = pd.read_csv('../../results/Halos_50_3_z1d25.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])

halo3_f_z0 = pd.read_csv('../../results/Halos_50_3_F_z0.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_f_z0d1 = pd.read_csv('../../results/Halos_50_3_F_z0d1.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_f_z0d24 = pd.read_csv('../../results/Halos_50_3_F_z0d24.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_f_z0d5 = pd.read_csv('../../results/Halos_50_3_F_z0d5.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_f_z0d7 = pd.read_csv('../../results/Halos_50_3_F_z0d7.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
halo3_f_z1d25 = pd.read_csv('../../results/Halos_50_3_F_z1d25.dat', sep='\s+', header=None, names=['X', 'Y', 'Z','Mvir','Rvir','Mdom','Npart'])
```

```python
#numero total de halos en z=0

print("Dark matter only halos at z=0:", len(halo3_d_z0))
print("Full physics halos at z=0:", len(halo3_f_z0))
```

```python
#cuantos halos tienen masa dominio igual a 0
n_d_0 = len(halo3_d_z0[halo3_d_z0['Mdom'] == 0])
n_f_0 = len(halo3_f_z0[halo3_f_z0['Mdom'] == 0])
print(f"Dark matter only halos with Mdom=0: {n_d_0}")
print(f"Full physics halos with Mdom=0: {n_f_0}")
```

```python
#distributions of Mdom of both simulations
def plot_mdom_distribution(halo_data_d, halo_data_f, z_label):
    plt.figure(figsize=(10,6))
    plt.hist(halo_data_d['Mdom'], bins=30, alpha=0.5, label='Dark Matter Only', color='blue', density=True)
    plt.hist(halo_data_f['Mdom'], bins=30, alpha=0.5, label='Full Physics', color='orange', density=True)
    plt.yscale('log')
    plt.xlabel('Dominant Mass (Mdom)')
    plt.ylabel('Density')
    plt.title(f'Distribution of Dominant Mass (Mdom) at {z_label}')
    plt.legend()
    plt.grid()
    plt.show()

plot_mdom_distribution(halo3_d_z0, halo3_f_z0, 'z=0')
```

```python
plot_mdom_distribution(halo3_d_z0d1, halo3_f_z0d1, 'z=0.1')
```

```python
plot_mdom_distribution(halo3_d_z0d24, halo3_f_z0d24, 'z=0.24')
```

```python
plot_mdom_distribution(halo3_d_z0d5, halo3_f_z0d5, 'z=0.5')
```

```python
plot_mdom_distribution(halo3_d_z0d7, halo3_f_z0d7, 'z=0.7')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

datasets = [
    (halo3_d_z0, halo3_f_z0, 'z=0'),
    (halo3_d_z0d1, halo3_f_z0d1, 'z=0.1'),
    (halo3_d_z0d24, halo3_f_z0d24, 'z=0.24'),
    (halo3_d_z0d5, halo3_f_z0d5, 'z=0.5'),
    (halo3_d_z0d7, halo3_f_z0d7, 'z=0.7'),
    (halo3_d_z1d25, halo3_f_z1d25, 'z=1.25')
]

for i, (data_d, data_f, label) in enumerate(datasets):
    ax = axes[i]
    ax.hist(data_d['Mdom'], bins=30, alpha=0.5, label='Dark Matter Only', color='blue', density=True)
    ax.hist(data_f['Mdom'], bins=30, alpha=0.5, label='Full Physics', color='orange', density=True)
    ax.set_yscale('log')
    ax.set_xlabel('Dominant Mass (Mdom)')
    ax.set_ylabel('Density')
    ax.set_title(f'{label}')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
