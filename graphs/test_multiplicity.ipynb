```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = "results/fof_disk_stats.txt"
#cols = HaloID N_subhalos_with_disk Mass_1e10_Msun_h
data = pd.read_csv(path, delim_whitespace='\s+', comment='#', names=['HaloID', 'N_subhalos_with_disk', 'Mass_1e10_Msun_h'])
```

```python
data['N_subhalos_with_disk'].sum()
```

```python
#graficar N_subhalos_with_disk vs Mass_1e10_Msun_h
plt.figure(figsize=(10,6))
plt.scatter(data['Mass_1e10_Msun_h'], data['N_subhalos_with_disk'], alpha=0.7)
plt.xlabel('Mass (10^10 Msun/h)')
plt.ylabel('Number of Subhalos with Disk')
#que los ticks del eje y sean enteros
plt.yticks(np.arange(0, data['N_subhalos_with_disk'].max()+1, 1))
plt.title('Number of Subhalos with Disk vs Halo Mass')
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()  
```

```python
#graficar suma cumulativa de N_subhalos_with_disk vs Mass_1e10_Msun_h
data_sorted = data.sort_values(by='Mass_1e10_Msun_h')
data_sorted['Cumulative_N_subhalos_with_disk'] = data_sorted['N_subhalos_with_disk'].cumsum()
plt.figure(figsize=(10,6))
plt.plot(data_sorted['Mass_1e10_Msun_h'], data_sorted['Cumulative_N_subhalos_with_disk'], marker='o')
plt.xscale('log')
plt.xlabel('Mass (10^10 Msun/h)')
plt.ylabel('Cumulative Number of Subhalos with Disk')
plt.title('Cumulative Number of Subhalos with Disk vs Halo Mass')
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()
```
